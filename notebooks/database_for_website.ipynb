```python
import pandas as pd

# Load the datasets
sites_df = pd.read_excel('/Users/kennyquintana/Desktop/merged_sites.xlsx', sheet_name='Sites')
incidents_df = pd.read_excel('/Users/kennyquintana/Desktop/merged_sites.xlsx', sheet_name='Incidents')
cross_sections_df = pd.read_excel('/Users/kennyquintana/Desktop/merged_sites.xlsx', sheet_name='CrossSectionsNHDNWM')

# 1. Calculate num_fatalities (incidents count per site)
# Group by site_id and count unique incidents
fatalities_count = incidents_df.groupby('site_id').size().reset_index(name='num_fatalities')

# Merge num_fatalities into sites_df
# Use 'left' join to keep all sites, filling missing values with 0
sites_df = pd.merge(sites_df, fatalities_count, on='site_id', how='left')
sites_df['num_fatalities'] = sites_df['num_fatalities'].fillna(0).astype(int)

# 2. Get Qmin and Qmax from CrossSectionsNHDNWM for xs_index == 3
# Filter for xs_index == 3 and select only relevant columns
xs_filtered = cross_sections_df[cross_sections_df['xs_index'] == 3][['site_id', 'Qmin', 'Qmax']]

# Merge Qmin and Qmax into sites_df
sites_df = pd.merge(sites_df, xs_filtered, on='site_id', how='left')

# Display the first few rows to verify the new columns
print(sites_df[['site_id', 'num_fatalities', 'Qmin', 'Qmax']].head())
sites_df.to_csv('/Users/kennyquintana/Desktop/lhd_website.csv', index=False)
```
